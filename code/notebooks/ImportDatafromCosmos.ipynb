```python
# Initialize list
dflist = []

db=client.get_database_client('enrichedCompanyData')
container=db.get_container_client('enrichedCompanyData')

for item in container.query_items(query='SELECT * FROM enrichedCompanyData',enable_cross_partition_query=True):
    print(json.dumps(item,indent=True))
    dflist.append(dict(item))

# Convert list to pandas DataFrame
df = pd.DataFrame(dflist)
```

### Flatten Nested JSON Documents to columns

```python
json_struct = json.loads(df.to_json(orient="records"))    
df_flat = pd.json_normalize(json_struct) #use pd.io.json
```

### Formatting Columns

```python
df_flat.rename(columns = {'leadershipInfo.gender':'leadergender'}, inplace = True)
df_flat.rename(columns = {'leadershipInfo.race':'leaderrace'}, inplace = True)

df_flat['leadergender'] = df_flat['leadergender'].str.lower()
df_flat['leaderrace'] = df_flat['leaderrace'].str.upper()
```

### Logic to define if a business is owned by Woman or not

```python
def myfuncwomenowned(x):
    if x == 'female':
        return 'YES'
    else:
        return 'NO'

df_flat['isWomanOwned'] = df_flat.apply(lambda x: myfuncwomenowned(x.leadergender),axis=1)
```

```python
def myfuncminorrace(x):
    if x == 'B_NL' :
        return 'Black or African-American'
    elif x == 'AI_AN':
        return 'American Indian or Alaska Native'
    elif x == 'A':
        return 'Asian'
    elif x == 'HL':
        return 'Hispanic or Latino'
    elif x == 'W_NL':
        return 'White or European American'
    else:
        return None

df_flat['leader_race_expanded'] = df_flat.apply(lambda x: myfuncminorrace(x.leaderrace),axis=1)
```
